```python
# What college major has the highest mid-career salary? How much do graduates with this major earn? (Mid-career is defined as having 10+ years of experience).
print(df.loc[df['Mid-Career Median Salary'].idxmax(), 'Undergraduate Major'])
print(int(df['Mid-Career Median Salary'].max()))
```

```python
# Which college major has the lowest starting salary and how much do graduates earn after university?
print(df.loc[df['Starting Median Salary'].idxmin(), 'Undergraduate Major'])
print(int(df['Starting Median Salary'].min()))
```

```python
# Which college major has the lowest mid-career salary and how much can people expect to earn with this degree?
print(df.loc[df['Mid-Career Median Salary'].idxmin(), 'Undergraduate Major'])
print(df.loc[df['Mid-Career Median Salary'].idxmin()])
```

```python
df['Spread'] = df['Mid-Career 90th Percentile Salary'] - df['Mid-Career 10th Percentile Salary']
df.head()
```

```python
# Find the top 5 degrees with the highest values in the 90th percentile. 
df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
# Find the degrees with the greatest spread in salaries. Which majors have the largest difference between high and low earners after graduation.
df.sort_values('Spread', ascending=False)[['Undergraduate Major', 'Spread']].head()
```
